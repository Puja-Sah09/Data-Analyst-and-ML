```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Example DataFrame (replace with your actual data)
data = {
    'Date': pd.date_range(start='2015-01-01', periods=200, freq='5D'),  # Example dates
    'Distance (km)': [10.2, 8.5, 12.1, 15.3, 9.7] * 40,
    'Average Heart Rate (bpm)': [150, 155, 148, 160, 145] * 40
}
df_run = pd.DataFrame(data)

# Convert 'Date' to datetime (if not already done)
df_run['Date'] = pd.to_datetime(df_run['Date'])

# Filter data for years 2015 to 2018 (inclusive)
runs_subset_2015_2018 = df_run[(df_run['Date'] >= '2015-01-01') & (df_run['Date'] <= '2018-12-31')]

# Resample to weekly sums and ensure continuous date range
runs_subset_2015_2018.set_index('Date', inplace=True)
df_run_dist_wkly = runs_subset_2015_2018['Distance (km)'].resample('W').sum().asfreq('W', fill_value=0)

# Decompose the time series
decomposed = sm.tsa.seasonal_decompose(df_run_dist_wkly, extrapolate_trend='freq', period=52)

# Create the plot
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(15, 10), sharex=True)

# Plot observed data
decomposed.observed.plot(ax=ax1, legend=False, color='blue')
ax1.set(ylabel='Observed', title='Time Series Decomposition of Weekly Running Distances')

# Plot trend component
decomposed.trend.plot(ax=ax2, legend=False, color='red')
ax2.set(ylabel='Trend')

# Plot seasonal component
decomposed.seasonal.plot(ax=ax3, legend=False, color='green')
ax3.set(ylabel='Seasonal')

# Plot residual component
decomposed.resid.plot(ax=ax4, legend=False, color='gray')
ax4.set(ylabel='Residual')
ax4.set_xlabel('Date')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data: generating random heart rates for demonstration
np.random.seed(0)
heart_rates = np.random.randint(90, 180, 100)  # Random heart rate data

# Define heart rate zones details
hr_zones = [100, 125, 133, 142, 151, 173]
zone_names = ['< 100', '100 - 124', '125 - 132', '133 - 141', '142 - 172', '> 172']
zone_colors = ['green', 'yellow', 'orange', 'tomato', 'red', 'maroon']

# Create plot
fig, ax = plt.subplots()

# Plot histogram
n, bins, patches = ax.hist(heart_rates, bins=hr_zones, alpha=0.5, edgecolor='black')

# Apply colors to each bin
for i in range(len(patches)):
    patches[i].set_facecolor(zone_colors[i])

# Customize the plot
ax.set(title='Distribution of Heart Rates', ylabel='Number of Occurrences', xlabel='Heart Rate')
ax.xaxis.set(ticks=hr_zones)

# Display the plot
plt.show()
```

```python
import pandas as pd

# Sample DataFrames for running, walking, and cycling
df_run = pd.DataFrame({
    'Type': ['Run'] * 5,
    'Distance (km)': [5, 10, 8, 15, 20],
    'Climb (m)': [100, 150, 200, 250, 300],
    'Average Speed (km/h)': [10, 9.5, 10.2, 9.8, 10.5]
})

df_walk = pd.DataFrame({
    'Type': ['Walk'] * 5,
    'Distance (km)': [3, 6, 4, 7, 5],
    'Climb (m)': [50, 60, 40, 70, 55],
    'Average Speed (km/h)': [5, 5.5, 5.2, 5.1, 5.3]
})

df_cycle = pd.DataFrame({
    'Type': ['Cycle'] * 5,
    'Distance (km)': [20, 25, 30, 35, 40],
    'Climb (m)': [300, 400, 350, 450, 500],
    'Average Speed (km/h)': [20, 22, 21, 23, 24]
})

# Concatenate DataFrames
df_run_walk_cycle = pd.concat([df_run, df_walk, df_cycle], ignore_index=True)

# Columns of interest
dist_climb_cols = ['Distance (km)', 'Climb (m)']
speed_col = ['Average Speed (km/h)']

# Calculate totals for each type
df_totals = df_run_walk_cycle.groupby('Type')[dist_climb_cols].sum().reset_index()

print('Totals for different training types:')
display(df_totals)

# Calculate summary statistics
df_summary = df_run_walk_cycle.groupby('Type')[dist_climb_cols + speed_col].describe()

# Combine totals with summary
for i in dist_climb_cols:
    df_summary[(i, 'total')] = df_totals.set_index('Type')[i]

print('Summary statistics for different training types:')
display(df_summary)

```
